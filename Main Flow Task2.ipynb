```python
columns_to_drop=['thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed', 'description']
data2=data1.drop(columns=columns_to_drop)
```

```python
data2.columns
```

```python
#plotting views vs likes
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.hist(data2['views'], bins=50, alpha=0.5, label='Views')
plt.hist(data2['likes'], bins=50, alpha=0.5, label='Likes')
plt.title('Views vs Likes (Histogram)') 
plt.xlabel('Count')
plt.ylabel('Frequency') 
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data2['views'], data2['likes'], alpha=0.5) 
plt.title('Views vs Likes (Scatter Plot)')
plt.xlabel('Views') 
plt.ylabel('Likes')
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='category_id', y='views', data=data2) 
plt.title('Box Plot of Views across Different Categories') 
plt.xlabel('Category ID') 
plt.ylabel('Views') 
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(data2['views'], bins=50, alpha=0.5, label='Views')
plt.hist(data2['dislikes'], bins=50, alpha=0.5, label='disikes')
plt.title('Views vs Dislikes (Histogram)') 
plt.xlabel('Count')
plt.ylabel('Frequency') 
plt.legend(loc='upper right')
plt.show()
```

```python
# numeric values only
numeric_data2=data2.select_dtypes(include=[float,int])
# Calculate the correlation matrix
correlation_matrix = numeric_data2.corr() 
# Plot the heatmap
plt.figure(figsize=(12,8)) 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5) 
plt.title('Correlation Matrix')
plt.show()
```
