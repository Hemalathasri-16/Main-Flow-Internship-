```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\bhima\\OneDrive\\Desktop\\Mian Flow\\USvideos.csv")
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of 'views'
plt.figure(figsize=(10, 6))
sns.histplot(df['views'], bins=50, kde=True)
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for 'views', 'likes', and 'dislikes'
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['views', 'likes', 'dislikes']])
plt.title('Boxplot of Views, Likes, and Dislikes')
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix for numeric columns only
correlation_matrix = numeric_df.corr()

# Generate a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Convert 'publish_time' to datetime
df['publish_time'] = pd.to_datetime(df['publish_time'])

# Extract year, month, and day
df['publish_year'] = df['publish_time'].dt.year
df['publish_month'] = df['publish_time'].dt.month
df['publish_day'] = df['publish_time'].dt.day

# Print the first few rows to check the new columns
print(df[['publish_time', 'publish_year', 'publish_month', 'publish_day']].head())
```

```python
# Bar plot for category_id
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category_id')
plt.title('Distribution of Category ID')
plt.xlabel('Category ID')
plt.ylabel('Count')
plt.show()
```
